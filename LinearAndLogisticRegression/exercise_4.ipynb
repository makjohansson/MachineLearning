 ## Exercise 4: Nonlinear logistic regression

```python
import numpy as np
import matplotlib.pyplot as plt
import utils
from matplotlib.colors import ListedColormap

X = np.loadtxt('A2_datasets_2020/microchips.csv', delimiter=',', usecols=(0,1))
y = np.loadtxt('A2_datasets_2020/microchips.csv', delimiter=',', usecols=(2))
```

 <ul>
     <li>Final learning rate and number of iterations</li>
     <li>Produce plot J(beta) as a function over iterations, print the learning rate and number of iterations</li>
     <li>Produce plot for the corresponding decision boundary and calculate the training errors</li>
    <li>Display plot for model with degree 2</li>
 </ul>

```python
degree = 2
learning_rate = 0.3
n_iterations = 500
utils.plot_cost(X, y, n_iterations, learning_rate, sub=True)
XX = utils.mapFeatures(X[:,0], X[:,1], degree)
XX_betas = utils.gradient_decent_2(XX, y, n_iterations, learning_rate)
utils.plotting_boundaries(X, XX_betas, y, degree)
XXX = utils.mapFeatures(X[:,0], X[:,1], degree)
p = utils.sigmoid(np.dot(XXX, XX_betas))
pp = np.round(p)
errors_train = np.sum(y!=pp)
plt.title('Training errors:' + str(errors_train))
plt.show()
```
