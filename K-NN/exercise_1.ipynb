## Exercise 1: k-NN Classification

```python
import numpy as np
import matplotlib.pyplot as plt
import A1

X = np.loadtxt('A1_datasets/microchips.csv', delimiter=',')
y = np.loadtxt('A1_datasets/microchips.csv', delimiter=',', usecols=2)

test = np.array([[-0.3, 1.0],
              [-0.5, -0.1],
              [0.6, 0.0]
            ])
```

<ul>
   <li>Plot the original microchip data using different markers for the two classes OK('o') and Fail('x').</li>
</ul>

```python
A1.plot_OK_FAIL(X, y)
```

<ul>
   <li>k-NN to predict whether the three unknown microchips(test) are likely to be OK or Fail</li>
</ul>

```python
k_s = [1,3,5,7]
for i in range(0, len(k_s)):
  k = A1.kNN(X, test, k=k_s[i])
  print(f'k = {k_s[i]}')
  A1.print_test_result(k, test)
```

<ul>
   <li>A 2x2 plot showing the decision boundary and the traning errors for k=1,3,5,7</li>
</ul>

```python
plt.figure(figsize=(12, 8))
for i in range(0, len(k_s)):
    yy = A1.kNN(X, X[:,:2], k=k_s[i])
    errors = np.where(y != yy)
    title = title = (f'k = {k_s[i]}, traning erros = {len(errors[0])}')
    plt.subplot(2,2,i+1).title.set_text(title)
    A1.plot_kNN(X, k_s[i], y)
```
