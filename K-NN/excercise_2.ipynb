## Exercise 2: k-NN Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import A1
from sklearn.metrics import mean_squared_error

data = np.loadtxt('A1_datasets/polynomial200.csv', delimiter=',')

X_train = data[:100]
X_test = data[100:]
```

<ul>
   <li>Plot of the training and test set side-by-side in a 1 × 2 pattern.</li>
</ul>

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.scatter(X_train[:, 0], X_train[:,1], s=10)
plt.subplot(1,2,2)
plt.scatter(X_test[:, 0], X_test[:,1], s=10)
plt.show()
```

<ul>
   <li>A 2 × 2 plot showing the k-NN regression result and the MSE training error for k = 1, 3, 5, 7.</li>
</ul>

```python
points = np.linspace(X_train[:,0].min(), X_train[:,0].max(), num=100)

k_s = [1,3,5,7]
plt.figure(figsize=(12, 8))
for i in range(0, len(k_s)):
    y_values = A1.get_y_values(X_train, X_train[:,0], k_s[i])
    Y_plot = A1.get_y_values(X_train, points, k_s[i] )
    XX = np.c_[points, Y_plot]
    MSE = np.round(np.square(np.subtract(X_train[:,1], y_values)).mean(), 2)
    title = (f'k = {k_s[i]}, MSE = {MSE}')
    plt.subplot(2,2,i+1).title.set_text(title)
    A1.plot_kNN_regression(X_train, XX)
plt.show()
```

<ul>
 <li>MSE test error for k = 1, 3, 5, 7</li>
</ul>

```python
for i in range(0, len(k_s)):
    y_values = A1.get_y_values(X_train, X_test[:, 0], k_s[i])
    MSE = np.round(np.square(np.subtract(X_test[:,1], y_values)).mean(), 2)
    print(f'Test MSE for k = {k_s[i]}, {MSE}')
```

K = 5 gives the best regression. The MSE for k=5 is the lowest which means that the distances between the points in the testset
and the points in the trained model are shorter than for the other k's.
